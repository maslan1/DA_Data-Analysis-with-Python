# Which brand had the largest absolute â‚¬ difference in average transaction value between domestic and international?

```python
df_new.sample(8)
```

```python
df_new["dom_int"] = np.where(df_new["buyer_country"] == df_new["seller_country"], "domestic", "international")
```

```python
df_new.sample(7)
```

```python
df_new["avg_dom"] = df_new[df_new["dom_int"] == "domestic"].groupby("brand").total.transform("mean")
df_new["avg_int"] = df_new[df_new["dom_int"] == "international"].groupby("brand").total.transform("mean")
df_new.tail()
```

```python
a = df_new.groupby("brand").avg_dom.mean()
b = df_new.groupby("brand").avg_int.mean()
a, b
```

```python
K = pd.DataFrame({"brand": a.index, "avg_dom": a.values, "avg_int":b.values, "diff":abs(a.values-b.values)})
K
```

```python
K.loc[K["diff"].idxmax()]
```
